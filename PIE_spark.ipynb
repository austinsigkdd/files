```
linesRDD = sc.textFile("/FileStore/tables/hw7nzjot1501531904757/wordcount.txt")
```

```
linesRDD.collect()
```

```
wordsRDD = linesRDD.flatMap(lambda line: line.split(" "))
```

```
wordIntCountRDD = wordsRDD.map(lambda word: (word, 1))
```

```
wordIntCountRDD.collect()
```

```
wordCountRDD = wordIntCountRDD.reduceByKey(lambda a, b : a + b)
```

```
wordCountRDD.collect()
```
